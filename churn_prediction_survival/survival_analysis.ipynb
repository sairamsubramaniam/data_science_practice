```python
from lifelines import KaplanMeierFitter
from lifelines.datasets import load_waltons
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
import numpy as np

df = load_waltons()
T = df["T"]
E = df["E"]

kmf = KaplanMeierFitter()
kmf.fit(T, event_observed=E)
kmf.plot()
```
